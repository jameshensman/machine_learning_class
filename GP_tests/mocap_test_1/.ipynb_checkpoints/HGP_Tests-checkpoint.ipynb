```python
%pylab inline
import numpy as np
import pylab as pb
import GPy
import urllib
```

```python
time_stamps = 100
x_parent = np.linspace(0, 2*np.pi, time_stamps)
x_sons = []
n_sons = 3
for i in range(0, n_sons):
    tmp = np.linspace(0, 2*np.pi, time_stamps)
    x_sons.append(tmp)

g_n = np.sin(x_parent)
fn_list = []
pb.plot(x_parent, g_n)
mu = 0
sigma = 0.01
for i in range(0, n_sons):
    tmp = np.sin(x_sons[i]) + np.random.normal(mu, sigma, x_sons[i].shape[0])
    fn_list.append(tmp)
    pb.plot(x_sons[i], tmp)
```

```python
sup_input = x_sons[0]
sup_input_id = np.zeros(sup_input.shape[0])
sup_output = fn_list[0]
for i in range(1,n_sons):
    sup_input = np.append(sup_input, x_sons[i])
    tmp_input_id = np.linspace(i,i,x_sons[i].shape[0])
    sup_input_id = np.append(sup_input_id, tmp_input_id)
    sup_output = np.append(sup_output, fn_list[i])

T_hat = np.vstack((sup_input, sup_input_id))
T_hat = np.transpose(T_hat)
Y_hat = sup_output[:,None]

k1 = GPy.kern.RBF(1, variance=0.1, lengthscale=100.) #+ GPy.kern.Matern32(1, variance=2., lengthscale=40.)
k2 = GPy.kern.RBF(1, variance=1., lengthscale=5.) #+ GPy.kern.Matern32(1, variance=2., lengthscale=40.)
kernels = [k1,k2]
HGP = GPy.kern._src.independent_outputs.Hierarchical(kernels)
model = GPy.models.GPRegression(T_hat, Y_hat, HGP)
model.optimize()
mu, var = model._raw_predict(T_hat)
for i in range(n_sons):
        model.plot(fignum=1,fixed_inputs=[(1, i)], plot_raw=True)

for i in range(0, n_sons):
    pb.plot(x_sons[i], fn_list[i])

print model
```

## Computing the parent posterior given its children

### Prior of the parent

```python
kernel_parent = model.kern.rbf_1
inp = np.arange(x_sons[0].shape[0])
T = inp[:, None]
C_parent = kernel_parent.K(T, None)
plt.imshow(C_parent)
K_HGP = np.matrix(C)
```

```python
C_parlist = []
for i in range(0,n_sons):
    C_parlist.append(C_parent)

tmp_kg = np.vstack(C_parlist)
full_K = np.vstack([np.hstack([C_parent, tmp_kg.T]), np.hstack([tmp_kg, K_HGP])])    
plt.figure(1)
plt.title('Full joint covariance')
plt.imshow(full_K)

posterior_mean = tmp_kg.T*np.linalg.inv(K_HGP)*Y_hat
posterior_cov = C_parent - (tmp_kg.T*np.linalg.inv(K_HGP)*tmp_kg)

plt.figure(2)
plt.title('Posterior covariance of parent')
plt.imshow(posterior_cov)
time_size = C_parent.shape[0]
```
